```python
pascal_voc_classes = [
    "background", "aeroplane", "bicycle", "bird", "boat", "bottle", 
    "bus", "car", "cat", "chair", "cow", "diningtable", "dog", 
    "horse", "motorbike", "person", "pottedplant", "sheep", "sofa", 
    "train", "tvmonitor"
]
```

```python

# Preprocess the input image
def preprocess_image(image_path):
    # Load image and resize to 256x256
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (256, 256))
    # Convert to tensor and normalize
    preprocess = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    return preprocess(image).unsqueeze(0), image  # Add batch dimension and return original image

# Perform inference and visualize
def segment_image(model, image_tensor, original_image):
    with torch.no_grad():
        # Predict the segmentation mask
        output = model(image_tensor)['out']
        output = output.squeeze(0).cpu().numpy()  # Remove batch dimension and convert to numpy array
        # Get the class with the highest probability at each pixel
        segmentation_map = np.argmax(output, axis=0)

    # Get unique class indices in the segmentation map
    unique_classes = np.unique(segmentation_map)

    # Print out the identified classes
    print("Classes present in the segmentation:")
    for class_idx in unique_classes:
        print(f"Class {class_idx}: {pascal_voc_classes[class_idx]}")
        
    # Visualize the original image and the segmentation mask
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(original_image)
    plt.title("Original Image")
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    # Colorize the segmentation map
    colors = np.array([[0, 0, 0], [0, 255, 0], [0, 0, 255], [255, 0, 0]])  # Add more colors if needed
    color_mask = colors[segmentation_map % len(colors)]
    plt.imshow(color_mask)
    plt.title("Segmentation Mask")
    plt.axis('off')
    
    plt.show()
    
    
```

```python
import torch
import numpy as np
import cv2
import matplotlib.pyplot as plt
from torchvision import models, transforms

# Load the pre-trained DeepLabV3 model with ResNet-50 backbone
model = models.segmentation.deeplabv3_resnet50(weights=models.segmentation.DeepLabV3_ResNet50_Weights.COCO_WITH_VOC_LABELS_V1)
model.eval()

# Define the preprocessing function
def preprocess_image(image_path):
    # Load the image using OpenCV and convert it to RGB
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # Resize the image for faster processing (DeepLabV3 can handle larger sizes if needed)
    image = cv2.resize(image, (512, 512))
    
    # Convert the image to tensor and normalize as expected by the model
    preprocess = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    
    return preprocess(image).unsqueeze(0), image  # Return tensor and original image

# Define the function to perform segmentation and visualize the results
def segment_image(model, image_tensor, original_image):
    with torch.no_grad():
        # Predict the segmentation mask
        output = model(image_tensor)['out']
        output = output.squeeze(0).cpu().numpy()  # Remove batch dimension and convert to numpy array
        # Get the class with the highest probability at each pixel
        segmentation_map = np.argmax(output, axis=0)

    # Define colors for each class (using a simple color palette)
    num_classes = output.shape[1]
    colors = np.random.randint(0, 255, (num_classes, 3), dtype=np.uint8)
    
    # Create a colorized mask based on the segmentation_map
    color_mask = colors[segmentation_map]
    
    # Display the original image and the segmentation map
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(original_image)
    plt.title("Original Image")
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(color_mask)
    plt.title("Segmentation Mask")
    plt.axis('off')
    
    plt.show()
```

```python
# Load and preprocess the image
image_path = "C:/Users/Baron/Desktop/EE_267_Repo/EE_267/%PATH_EE267%/EE267_env//pictures/ny.jpg" # Replace with the path to your image
image_tensor, original_image = preprocess_image(image_path)

# Perform segmentation and display the results
segment_image(model, image_tensor, original_image)
```
