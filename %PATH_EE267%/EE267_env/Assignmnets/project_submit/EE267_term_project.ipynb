### 1. Extraction of images from the video dataset.
    Run this function to convert the video dataset to run the extraction

```python
### Extract frames function
def extract_frames(video_path, output_folder):
    # Extract the video file name without extension to use as a prefix
    video_name = os.path.splitext(os.path.basename(video_path))[0]

    # Check if the output folder exists, and if so, clear it
    if os.path.exists(output_folder):
        shutil.rmtree(output_folder)  # Remove all existing files in the folder
    os.makedirs(output_folder)  # Re-create the empty folder

    # Open the video file
    video_capture = cv2.VideoCapture(video_path)
    
    # Check if the video was opened successfully
    if not video_capture.isOpened():
        print("Error: Could not open video.")
        return

    frame_count = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_rate = int(video_capture.get(cv2.CAP_PROP_FPS))
    print(f"Total frames: {frame_count}, Frame rate: {frame_rate} fps")

    # Extract and save frames
    frame_index = 0
    while True:
        ret, frame = video_capture.read()
        if not ret:
            break
        
        # Define the filename for each frame with the video name as a prefix
        frame_filename = os.path.join(output_folder, f"{video_name}_frame_{frame_index:05d}.jpg")
        
        # Write the frame to the output folder
        cv2.imwrite(frame_filename, frame)
        
        print(f"Extracted frame {frame_index}")
        frame_index += 1

    # Release the video capture object
    video_capture.release()
    print(f"Frames saved to {output_folder}")
```

```python
import cv2
import numpy as np
from PIL import Image
```

```python
#===========================================
""" 
Enhancements: 
    Preprocess Image: Processes an input image to enhance its contrast by applying histogram equalization to the luminance channel
                      The enhancement of contrast helps make features in the image more distinct, improving performance in 
                      downstream tasks like edge detection, segmentation, or feature extraction
    Gamma Correction: Nonlinear operation used to adjust the luminance or brightness levels of an image. It is applied to compensate for the way 
                      human eyes perceive light and the way devices like monitors or cameras display or capture light intensity.
    Crop Image : This function crops the image so that the swimmer is in size or in "view" for the detection algorithm.
"""
#===========================================
def preprocess_image(image):
    # Convert image to YUV color space
    image_yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
    # Apply histogram equalization on the Y channel (luminance)
    image_yuv[:, :, 0] = cv2.equalizeHist(image_yuv[:, :, 0])
    # Convert back to BGR color space
    processed_image = cv2.cvtColor(image_yuv, cv2.COLOR_YUV2BGR)
    return processed_image

def gamma_correction(image, gamma=2):
    # Build a lookup table mapping pixel values [0, 255] to adjusted gamma values
    inv_gamma = 1.0 / gamma
    table = np.array([(i / 255.0) ** inv_gamma * 255 for i in np.arange(256)]).astype("uint8")
    # Apply gamma correction
    gamma_corrected_image = cv2.LUT(image, table)
    
    # Show the image after gamma correction
    ##cv2.imshow("Gamma Corrected Image", gamma_corrected_image)
    ##print("Showing Gamma Corrected Image...")
    ##cv2.waitKey(0)  # Wait for a key press
    ##cv2.destroyAllWindows()
    return gamma_corrected_image


def crop_image(image):
    # Get current dimensions
    height, width = image.shape[:2]
    print(f"Original size: {width}x{height}")

    # Calculate cropping box (center crop)
    crop_width = width // 2
    crop_height = height // 2

    start_x = (width - crop_width) // 2
    start_y = (height - crop_height) // 2
    end_x = start_x + crop_width
    end_y = start_y + crop_height

    # Perform the crop
    cropped_image = image[start_y:end_y, start_x:end_x]
    print(f"Cropped image size: {crop_width}x{crop_height}")
    return cropped_image
    


def enhance_image(image): # Change back to image
#def enhance_image(input_path): # Change back to image
    # Apply histogram equalization
    #image = cv2.imread(input_path)
    
    image = preprocess_image(image)
    image = gamma_correction(image, gamma=3)

    image = crop_image(image)

    
    # Uncomment for debugging purposes. Shows the final enhanced image one at a time.
    #cv2.imshow("Enhanced Image", image)
    #print("Showing Enhanced Image...")
    #cv2.waitKey(0)  # Wait for a key press
    #cv2.destroyAllWindows()
    return image

#Uncomment as needed for debugging Purposes
def enhance_image_SHOW(input_path): # Change back to image
    image = cv2.imread(input_path)
    image = preprocess_image(image)
    image = gamma_correction(image, gamma=3)
    image = crop_image(image)

    
    #Uncomment for debugging purposes. Shows the final enhanced image one at a time.
    cv2.imshow("Enhanced Image", image)
    print("Showing Enhanced Image...")
    cv2.waitKey(0)  # Wait for a key press
    cv2.destroyAllWindows()
    return image

```

```python
# Function to process and display images one at a time
def process_and_display_images(input_folder, output_folder):
    # Create the output folder if it doesn't exist
    
    # Check if the output folder exists, and if so, clear it
    if os.path.exists(output_folder):
        shutil.rmtree(output_folder)  # Remove all existing files in the folder
    os.makedirs(output_folder)  # Re-create the empty folder

   # Loop through all files in the input folder
    for filename in os.listdir(input_folder):
        # Construct full file path
        input_path = os.path.join(input_folder, filename)
        
        # Extract the file name and extension
        file_base, file_ext = os.path.splitext(filename)
        output_filename = f"{file_base}_pp{file_ext}"  # Add the '_pp' suffix
        output_path = os.path.join(output_folder, output_filename)

        # Check if it is a file and has an image extension
        if os.path.isfile(input_path) and filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
            # Read the image
            image = cv2.imread(input_path)
            if image is not None:
                # Enhance the image
                enhanced_image = enhance_image(image)
                
                # Save the enhanced image to the output folder
                cv2.imwrite(output_path, enhanced_image)
                print(f"Processed and saved: {output_path}")
            else:
                print(f"Failed to read image: {input_path}")
```

### MediaPipe Object Detection
    1. Before Meta's Segment Anything Model(SAM) is used, the enhanced images an object detection module must be used to located the swimmer. 
    
    2. This is due to the nature of the SAM being coordinate based. When the user provides a coordinate of where the object is, the SAM performs well of segmenting the object that is around that coordinate.
    
    3. The more coordinates that are provided for that singular object, or multiple objects, the better segmentation is yeilded.

```python
import cv2
import mediapipe as mp
import numpy as np

# Initialize MediaPipe BlazePose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=True, model_complexity=1, enable_segmentation=True , min_detection_confidence=0.3, min_tracking_confidence=0.2)

def detect_swimmer_with_blazepose(image_path, debugging=True):
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Failed to load image {image_path}")
        return None, None

    # Create a copy of the image for visualization (to avoid modifying the original image)
    visual_image = image.copy()

    # Convert the image to RGB (MediaPipe requires RGB input)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Process the image with BlazePose
    results = pose.process(image_rgb)

    # Get image dimensions
    h, w, _ = image.shape

    # Draw grid for visualization (optional)
    grid_color = (255, 255, 255)  # White color for the grid
    num_rows, num_cols = 10, 10  # Define grid dimensions
    for row in range(1, num_rows):
        y = row * h // num_rows
        cv2.line(visual_image, (0, y), (w, y), grid_color, 1)
    for col in range(1, num_cols):
        x = col * w // num_cols
        cv2.line(visual_image, (x, 0), (x, h), grid_color, 1)

    # List to store detected keypoints
    detected_points = []

    # Extract landmarks (key points) if they exist
    if results.pose_landmarks:
        for idx, landmark in enumerate(results.pose_landmarks.landmark):
            # Convert normalized coordinates to pixel coordinates
            x = int(landmark.x * w)
            y = int(landmark.y * h)
            confidence = landmark.visibility

            # Append the keypoint (x, y) to the list
            detected_points.append([x, y])

            # Visualize the keypoints on the visualization image
            if confidence > 0.5:  # Only draw points with good visibility
                cv2.circle(visual_image, (x, y), 5, (0, 255, 0), -1)  # Green dots for keypoints
                cv2.putText(visual_image, str(idx), (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 255, 0), 1)
        
    # Handle the case when no points are detected
    if not detected_points:
        print("No swimmer detected.")
        return None, visual_image
    if (debugging == True):
        return np.array(detected_points), visual_image  # Return unmodified original image
    else:
        return np.array(detected_points), image 
# Initialize MediaPipe BlazePose
```

### Swimmer Segmentation Using Meta's Segment Anything Model (SAM). 
    The use of SAM in this project is the coordinate based implementation. Coordinates are provided by the MediaPipe framework which are 
    fed into the SAM model and the segmentation occurs

```python
import torch
from torchvision import models, transforms
import torchvision
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
from segment_anything import sam_model_registry, SamPredictor
```

```python
def show_mask(mask, ax, random_color=False):
    if random_color:
        color = np.concatenate([np.random.random(3), np.array([0.6])], axis=0)
    else:
        color = np.array([30/255, 144/255, 255/255, 0.6])
    h, w = mask.shape[-2:]
    mask_image = mask.reshape(h, w, 1) * color.reshape(1, 1, -1)
    ax.imshow(mask_image)
    
def show_points(coords, labels, ax, marker_size=375):
    pos_points = coords[labels==1]
    neg_points = coords[labels==0]
    ax.scatter(pos_points[:, 0], pos_points[:, 1], color='green', marker='*', s=marker_size, edgecolor='white', linewidth=1.25)
    ax.scatter(neg_points[:, 0], neg_points[:, 1], color='red', marker='*', s=marker_size, edgecolor='white', linewidth=1.25)   
    
def show_box(box, ax):
    x0, y0 = box[0], box[1]
    w, h = box[2] - box[0], box[3] - box[1]
    ax.add_patch(plt.Rectangle((x0, y0), w, h, edgecolor='green', facecolor=(0,0,0,0), lw=2))
```

```python
def detect_and_segment_swimmer(image_path, model_checkpoint, model_type="vit_h", device="cpu"):


    # Detect swimmer using BlazePose
    detected_points, detected_image = detect_swimmer_with_blazepose(image_path, debugging=False)

    if detected_points is None:
        print("No swimmer detected.")
        return

    # Initialize SAM model
    sam = sam_model_registry[model_type](checkpoint=model_checkpoint)
    sam.to(device=device)
    predictor = SamPredictor(sam)

    # Set SAM's input image
    predictor.set_image(detected_image)

    # Use detected swimmer points as input for SAM
    input_point = detected_points
    input_label = np.array([1] * len(detected_points))  # Positive label for all points

    # Perform SAM prediction
    masks, scores, logits = predictor.predict(
        point_coords=input_point,
        point_labels=input_label,
        multimask_output=False,
    )

    # Visualize the results
    for i, (mask, score) in enumerate(zip(masks, scores)):
        plt.figure(figsize=(10, 10))
        plt.imshow(detected_image)
        show_mask(mask, plt.gca())
        show_points(input_point, input_label, plt.gca())
        plt.title(f"Mask {i + 1}, Score: {score:.3f}", fontsize=18)
        plt.axis('off')
        plt.show()
```

### Swimmer Segmentation Applied o all images in folder and saved in New Folder

```python
import torch
from torchvision import models, transforms
import torchvision
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
from segment_anything import sam_model_registry, SamPredictor
```

```python
def segment_swimmer_images(input_folder, output_folder, sam_checkpoint, model_type="vit_h", device="cpu"):
    """
    Processes images in the input folder, applies swimmer detection and segmentation,
    and saves the results in the output folder.

    Parameters:
    - input_folder: Folder containing input images.
    - output_folder: Folder to save the segmented images.
    - sam_checkpoint: Path to the SAM model checkpoint.
    - model_type: SAM model type (default: "vit_h").
    - device: Device to run the SAM model on (default: "cpu").
    """
    # Load SAM model for segmentation
    sam = sam_model_registry[model_type](checkpoint=sam_checkpoint)
    sam.to(device=device)
    predictor = SamPredictor(sam)
    
    # Check if the output folder exists, and if so, clear it
    if os.path.exists(output_folder):
        shutil.rmtree(output_folder)  # Remove all existing files in the folder
    os.makedirs(output_folder)  # Re-create the empty folder
    
    for filename in os.listdir(input_folder):
        image_path = os.path.join(input_folder, filename)
        
        # Detect swimmer in the image using BlazePose
        detected_points, image = detect_swimmer_with_blazepose(image_path, debugging=False)

        
        if detected_points is None or image is None:
            print(f"No swimmer detected in {filename}. Skipping...")
            continue  # Skip this image if no swimmer detected
        
        # Set SAM's input image
        predictor.set_image(image)
    
        # Use detected swimmer points as input for SAM
        input_point = detected_points
        input_label = np.array([1] * len(detected_points))  # Positive label for all points

        
        
        # Convert to RGB for SAM model
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        predictor.set_image(image_rgb)
        
        # Use SAM for segmentation with multimask_output set to False
        masks, scores, logits = predictor.predict(
            point_coords=input_point,
            point_labels=input_label,
            multimask_output=False  # Produce multiple masks
        )

        # Apply the mask to the image
        mask = masks[0]  # Retrieve the single mask
        mask_image = image_rgb * mask[..., np.newaxis]
        mask_image_bgr = cv2.cvtColor(mask_image, cv2.COLOR_RGB2BGR)
        
        # Save the segmented image
        output_filename = f"{os.path.splitext(filename)[0]}_segmented.jpg"
        output_path = os.path.join(output_folder, output_filename)
        cv2.imwrite(output_path, mask_image_bgr)
        print(f"Saved segmented image {output_filename}")
        
```
